```python
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

## 3. Training model

```python
YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 32

model.train(
    data=YAML_PATH,
    epochs=EPOCHS,
    batch=BATCH_SIZE,
    imgsz=IMG_SIZE
)
```

## 4. Model evaluation

```python
TRAINED_MODEL_PATH = 'runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(
    data=YAML_PATH,
    imgsz=IMG_SIZE,
    split='test'
)
```

## 5. Prediction

```python
IMG_PATH = '/content/safety_helmet_dataset/test/images/helmet-348-_jpg.rf.0742694df03cce9b44cdcf5610fd7cc0.jpg'
result = model.predict(IMG_PATH)[0]
result.plot()
```
